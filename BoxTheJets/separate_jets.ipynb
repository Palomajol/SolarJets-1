```python
import matplotlib.pyplot as plt
%matplotlib inline
from aggregation import Aggregator
from astropy.io import ascii
from astropy.table import Table
import numpy as np
plt.style.use('default')
```

```python
aggregator        = Aggregator('reductions/point_reducer_hdbscan_box_the_jets.csv', 
                               'reductions/shape_reducer_hdbscan_box_the_jets.csv')
```

```python
multijet_subjects = ascii.read('multiplejets_20220209.txt', format='fast_no_header')
```

```python
for subject in multijet_subjects:
    subject = subject[0]
    data_T1, clusters_T1 = aggregator.get_box_data(subject, 'T1')
    _, _, box_iou_T1     = aggregator.get_cluster_confidence(subject, 'T1')
    print(box_iou_T1)
    data_T5, clusters_T5 = aggregator.get_box_data(subject, 'T5')
    _, _, box_iou_T5     = aggregator.get_cluster_confidence(subject, 'T5')
    print(box_iou_T5)
    aggregator.find_unique_jets(subject, clip_iou=0.3)
    
    aggregator.plot_subject_both(subject)
```
